```python
import time
import numpy as np
import torch
from torch.utils.data import TensorDataset, DataLoader
from matplotlib import pyplot as plt
```

```python
device = torch.device('cuda')
learning_rate = 1e-2

x = torch.randn(64, 1000, device = device) # 64*1000
y = torch.randn(64, 10, device = device) # 64*10
loader = DataLoader(TensorDataset(x,y), batch_size = 8)

class Net(torch.nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = torch.nn.Linear(1000, 100)
        self.fc2 = torch.nn.Linear(100, 10)
                
    def forward(self, x):
        h = self.fc1(x)
        h_relu = torch.nn.functional.relu(h)
        y_pred = self.fc2(h_relu)
        return y_pred
                
model = Net()
model.to(device)

loss_list = []
                
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
                
for epochs in range(300):
    loss_sum = 0
    for x_batch, y_batch in loader:
        y_pred = model(x_batch)
        loss = torch.nn.functional.mse_loss(y_pred, y_batch)
        
        loss_sum += loss.item()
        print(loss.item())
        optimizer.step()
        optimizer.zero_grad()
    loss_sum /= 8.0
    loss_list.append(loss_sum)
                
```

```python
from matplotlib import pyplot as plt
plt.plot(loss_list)
plt.ylim((1.00, 1.01))
```

# Train a model to identify the quadrant of the point

```python
device = torch.device('cuda')
# Hyperparameters
size = 5000
test_size = 500
batch_size = 100
learning_rate = 1e-2
momentum = 0.9
HIDDEN_CHANNEL = 200
OUT_CHANNEL = 4
epoch = 100
```

```python
# Build training data
x_train = torch.randn(size, 2, device = device)
y_train = torch.zeros(size, 4, device = device, dtype = torch.long)
for i in range(x_train.shape[0]):
    if x_train[i][0] > 0 and x_train[i][1] > 0:
        y_train[i][0] = 1 # first quadrant
    elif x_train[i][0] < 0 and x_train[i][1] > 0:
        y_train[i][1] = 1 # second quadrant
    elif x_train[i][0] < 0 and x_train[i][1] < 0:
        y_train[i][2] = 1 # third quadrant
    else:
        y_train[i][3] = 1 # fourth quadrant
classes = ('1', '2', '3', '4')
```

```python
train_loader = DataLoader(TensorDataset(x_train, y_train), batch_size = batch_size)
```

```python
def viz_data(x_train, y_train):
    for i in range(x_train.shape[0]):
        if y_train[i][0].to('cpu') == 1:
            plt.plot(x_train[i][0].to('cpu'), x_train[i][1].to('cpu'), 'ro')
        elif y_train[i][1].to('cpu') == 1:
            plt.plot(x_train[i][0].to('cpu'), x_train[i][1].to('cpu'), 'bo')
        elif y_train[i][2].to('cpu') == 1:
            plt.plot(x_train[i][0].to('cpu'), x_train[i][1].to('cpu'), 'go')
        else:
            plt.plot(x_train[i][0].to('cpu'), x_train[i][1].to('cpu'), 'yo')
    x_zero = np.zeros([100])
    y = np.linspace(4, -4, 100)
    x = np.linspace(4, -4, 100)
    y_zero = np.zeros([100])
    plt.plot(x_zero, y, 'k--')
    plt.plot(x, y_zero, 'k--')
```

```python
viz_data(x_train, y_train)
```

```python
class QuatNet(torch.nn.Module):
    def __init__(self):
        super(QuatNet, self).__init__()
        self.fc1 = torch.nn.Linear(2, HIDDEN_CHANNEL)
        self.fc2 = torch.nn.Linear(HIDDEN_CHANNEL, HIDDEN_CHANNEL)
        self.fc3 = torch.nn.Linear(HIDDEN_CHANNEL, OUT_CHANNEL)
                
    def forward(self, x):
        out = self.fc1(x)
        out = torch.nn.functional.relu(out)
        out = self.fc2(out)
        out = torch.nn.functional.relu(out)
        y_pred = self.fc3(out)
        y_pred = torch.nn.functional.softmax(y_pred)
        return y_pred
    
    
```

```python
model = QuatNet()
model.to(device)

#Train
loss_list = []

criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate, momentum = momentum)

total_time = 0
for epochs in range(epoch):
    ts = time.time()
    for x_batch, y_batch in train_loader:
        y_pred = model(x_batch)

        optimizer.zero_grad()
        #print(torch.max(y_batch, 1)[1])
        loss = criterion(y_pred, torch.max(y_batch, 1)[1])
        loss.backward()
        #print(loss.item())
    
        optimizer.step()
        
        loss_list.append(loss.item())
        epoch_time = time.time() - ts
        print("Epoch: {} with loss: {:.2f}, execution time: {:.2f}".format(epochs+1, loss.item(), epoch_time))
    total_time += time.time() - ts
print("Training process completed with {:.2f}".format(total_time))
```

```python
plt.plot(loss_list)
plt.title("Loss curve with \nepoch: {}, batch size: {}, lr: {}, momentum: {}".format(
            epoch, batch_size, learning_rate, momentum))
```

```python
x_test = torch.randn(test_size, 2, device = device)
y_test = torch.zeros(test_size, 4, device = device, dtype = torch.long)
y_groundtruth = torch.zeros(test_size, 4, device = device)

for i in range(test_size):
    if x_test[i][0] > 0 and x_test[i][1] > 0: 
        y_groundtruth[i][0] = 1 # First
    elif x_test[i][0] < 0 and x_test[i][1] > 0:
        y_groundtruth[i][1] = 1 # Second
    elif x_test[i][0] < 0 and x_test[i][1] < 0:
        y_groundtruth[i][2] = 1 # Third
    else:
        y_groundtruth[i][3] = 1 # Fourth
```

```python
out = model(x_test)
for i in range(test_size):
    _, predicted = torch.max(out[i], -1)
    if predicted == 0:
        y_test[i][0] = 1
    elif predicted == 1:
        y_test[i][1] = 1
    elif predicted == 2:
        y_test[i][2] = 1
    else:
        y_test[i][3] = 1
```

```python
viz_data(x_test, y_test)
```

```python
correct = 0
for i in range(test_size):
    if torch.all(torch.eq(y_groundtruth[i].long(), y_test[i].long())):
        correct += 1
        
print("Accuracy: {:.2f}".format(correct/test_size*100.))
```
